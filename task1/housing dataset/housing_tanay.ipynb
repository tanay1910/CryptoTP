```python
# Visualizing missing values
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
# Histograms for numerical columns
df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#selecting numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])
print(numerical_df.columns)
```

```python
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = numerical_df.corr() #numerical_df is the numerical dataframe


# Display the correlation matrix
print(corr_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
corr_matrix[corr_matrix > 0.5]
#print the pairs which have correlation greater than 0.5
print(corr_matrix[corr_matrix > 0.5])
# Visualizing the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix[corr_matrix > 0.5], annot=True, cmap='coolwarm')
plt.show()
#blue means positive correlation, red means negative correlation
```

```python
# Set a threshold
threshold = 0.7

# Iterate through the correlation matrix and print pairs with correlation > threshold
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold: #abs means both positive and negative correlation
            print(f"Correlation between {corr_matrix.columns[i]} and {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]:.2f}")
```

```python
# Set a threshold
threshold = -0.2

# Iterate through the correlation matrix and print pairs with correlation > threshold
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if (corr_matrix.iloc[i, j]) < threshold:
            print(f"Correlation between {corr_matrix.columns[i]} and {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]:.2f}")
```

```python
y = df['SalePrice']
print(y.describe())
```

```python
#find the columns with a r value of 0.5 or higher with SalePrice
corr_matrix['SalePrice'][corr_matrix['SalePrice'] > 0.5]
```

```python
y = df['SalePrice']
X = df[['OverallQual']]
print(X.head())
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
predictions = model.predict(X)
```

```python
#calculate rmse
from sklearn.metrics import mean_squared_error
import numpy as np
# Calculate the RMSE
rmse = np.sqrt(mean_squared_error(y, predictions))
print(rmse)
```

```python
#find the accuracy of the model
model.score(X, y)
```

```python
#prrint mean of sales price
print(y.mean())
```
