```python
import pandas as pd
```

```python
complete_df = pd.read_csv('./train.csv')
```

```python
complete_df.describe()
```

### Separating numerical and catergorical data

```python
numerical_df = complete_df.select_dtypes(include=['int64', 'float64'])
numerical_df.drop('SalePrice', axis=1, inplace=True)
numerical_df.head()
```

```python
category_df = complete_df.select_dtypes(include=['object'])
category_df.head()
```

### Handling null values in numerical data

```python
#show the columns with missing values in nu
missing_values = numerical_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
#fill the missing values with the mean of the column for numerical data
numerical_df = numerical_df.fillna(numerical_df.mean())
```

```python
missing_values = numerical_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

which means there are no null values in numerical df now

### Handling missing values in categorical data

```python
#show the colunms with missing values in category_df
missing_values = category_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
#drop the columns in category_df with a lot of missing values
category_df = category_df.drop(['Alley', 'MasVnrType', 'PoolQC','Fence','MiscFeature','FireplaceQu'], axis=1)
```

```python
#show the columns with missing values in category_df
missing_values = category_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
#fill the missing values with the mode of the column for category data
category_df = category_df.fillna(category_df.mode().iloc[0])
```

```python
numerical_df.head()
```

```python
category_df.head()
```

### Encoding the categorical data

```python
#do one hot encoding for the category data using SciKit Learn
from sklearn.preprocessing import OneHotEncoder

# One-hot encoding
encoder = OneHotEncoder(sparse=False)
encoded_array = encoder.fit_transform(category_df)

# Convert the array back to a DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(category_df.columns))
print(encoded_df)
category_df = encoded_df

```

### Normalising the df

```python
#normalise the numerical data using SciKit Learn
from sklearn.preprocessing import MinMaxScaler
# Normalise the data
scaler = MinMaxScaler()
normalised_array = scaler.fit_transform(numerical_df)
# Convert the array back to a DataFrame
normalised_df = pd.DataFrame(normalised_array, columns=numerical_df.columns)
print(normalised_df)
numerical_df = normalised_df
```

### Deciding the variables


```python
y  = complete_df['SalePrice']
X = pd.concat([numerical_df, category_df], axis=1)
#remove the target column from the X
```

```python
#show the correlation between the features and the target
correlation = X.corrwith(y)
correlation = correlation.sort_values(ascending=False)
print(correlation)
```

Including featrues with high correlation

```python
#modify the features to only include the features with a correlation of 0.3 or higher
# X = X[correlation[correlation > 0.3].index]
```

```python
#fit the data into a linear regression model dont do test train split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Create the model
model = LinearRegression()
# Fit the model
model.fit(X, y)
```

Taking in Test Data

```python
#now take the test data and use it to predict the target
test_df = pd.read_csv('./test.csv')
```

Managing null values in test data

```python
numerical_test_df = test_df.select_dtypes(include=['int64', 'float64'])
numerical_test_df = numerical_test_df.fillna(numerical_test_df.mean())
```

```python
missing_values = numerical_test_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

### Handling missing values in categorical data

```python
#select category data from test_df
category_test_df = test_df.select_dtypes(include=['object'])
category_test_df = category_test_df.drop(['Alley', 'MasVnrType', 'PoolQC','Fence','MiscFeature','FireplaceQu'], axis=1)
```

```python
#show the columns with missing values in category_df
missing_values = category_test_df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
#fill the missing values with the mode of the column for category data
category_test_df = category_test_df.fillna(category_test_df.mode().iloc[0])
```

```python
#ecnode the category data using the encoder from before
encoded_array = encoder.transform(category_test_df)
# Convert the array back to a DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(category_test_df.columns))

category_test_df = encoded_df
```

```python
#normalise the numerical data using SciKit Learn
# Normalise the data
normalised_array = scaler.transform(numerical_test_df)
# Convert the array back to a DataFrame
normalised_df = pd.DataFrame(normalised_array, columns=numerical_test_df.columns)

numerical_test_df = normalised_df
```

```python
X_test = pd.concat([numerical_test_df, category_test_df], axis=1)
```

```python
# get the predictions from the model
predictions = model.predict(X_test)
```

```python
# print the predictions
print(predictions[:5])
```

```python
#save the predictions to a csv file with id, SalePrice and make sure they have equal number of rows
test_df['SalePrice'] = predictions
test_df['Id'] = test_df.index
test_df = test_df[['Id', 'SalePrice']]
test_df.to_csv('predictions.csv', index=False)
```
